```python
import numpy as np
import h5py as hdf
import matplotlib as mpl
import matplotlib.pyplot as plt
import pickle


#class Field:
#    def __init__(self, name):
#        self.name = name
        
#class padField(Field):
    
#class 

class stratFile:
    def __init__(self, filepath):
        self.path = filepath
        self.pickled = False
        self.checkForPickleDump()
        self.waveFields = {}         # z, x, y
        self.wavePaddedFields = {}   # z, x, y
        self.waveFilteredFields = {}
        self.realFields = {}
        self.realPaddedFields = {}   # z, x, y
        self.realFilteredFields = {}
        self.processHDF5()

    def checkForPickleDump(self):
        self.pickled = False
        
    def processHDF5(self):
        hdfData = hdf.File(self.path,"r")
        self.numPts = [hdfData.get('nx')[()], hdfData.get('ny')[()], hdfData.get('nz')[()]]
        self.time = hdfData.get('time')[()]
        self.yScale = hdfData.get('LMFy')[()]
        self.visc = hdfData.get('nu')[()]
        self.diff = hdfData.get('diff')[()]
        self.grav = hdfData.get('grav')[()]
        self.gradv = hdfData.get('gradv')[()]
        self.step = hdfData.get('timestep')[()]
        
        bruVais = str(int(np.sqrt(self.grav)))
        self.pickleName = "N" + bruVais + "-" + str(int(self.numPts[0])) + str(int(self.numPts[1])) \
                              + "-" + str(int(self.step)) + ".p"

        self.processHDFFields(hdfData)
        
    def processHDFFields(self, hdfData):
        # Something that checks for a pickle dump
        # self.pickled = search directory for pickle file for 
        # if (self.pickled):
            # read it in
        # else:
            self.readHDFFields(hdfData)
            #self.generateInitialRealFields()
            self.generateInitialRealPaddedFields()
            
    def readHDFFields(self, hdfData, fields=['rho','u','v','w']):
        for field in fields:
            fieldData = hdfData[field][()]
            # Poongback uses real symmetry in the x direction and doesn't store nyquist mode
            # So this will be nz-nx-n dimensions
            nz,nx,n,_ = fieldData.shape
            # The numpy rfft and irfft routines use the nyquist mode so I need to create a 
            # 0-valued mode there, hence the nx+1
            wavespace = np.zeros((nz,nx+1,n), dtype=complex)
            # Only read in up to :nx, since that is all that is stored by Poongback
            wavespace[:,:nx,:].real = fieldData[:,:,:,0]
            wavespace[:,:nx,:].imag = fieldData[:,:,:,1]
            name = field + 'hat'
            # This will now be a wave field stored for use with np.fft and exploiting real conjugate sym
            # Thus its dimensions will be nz-(nx+1)-n
            self.waveFields[name] = wavespace   

    def generateRealFields(self, fields=['rhohat','uhat','vhat','what']):
        for key in fields:
            name = key[:-3]    # drop the _hat from the waveField name
            self.realFields[name] = self.convertToReal(self.waveFields[key])
            
    def generateInitialRealPaddedFields(self):
        for key, waveField in self.waveFields.items():
            name = key[:-3]    # drop the _hat from the waveField name
            paddedField = self.padField(waveField)
            self.realPaddedFields[name] = self.convertToReal(paddedField, pad=True)
            #self.realPaddedFields[name] = self.convertToRealPad(waveField)

    def padField(self, waveField):
        (nz,nx,n) = waveField.shape
        shiftField = np.fft.fftshift(waveField, axes=(0,2,))                    # Shift for easy padding
        shiftField[0,:,:] = 0.0                                                 # Zero out nyquist mode
        shiftField[:,:,0] = 0.0                                                 # Zero out nyquist mode
        padShiftField = np.pad(shiftField, ((int(nz/4), int(nz/4)),(0,int((nx-1)/2)),(int(n/4), int(n/4)))) # Pad zeros at front and back
        padField = np.fft.ifftshift(padShiftField, axes=(0,2,))                 # Shift back
        
        return padField
            
    def convertToReal(self, waveField, pad=False):
        (nz,nx,n) = waveField.shape
        #if (pad):
        #    (nz,nx,n) = waveField.shape
        #    nz = int(nz*2/3)
        #    n = int(n*2/3)
        #    nx = int((nx-1)*2/3 + 1)
        #print (nz, nx, n)
        real = np.fft.irfft(np.fft.ifft(np.fft.ifft(waveField, axis=0), axis=2), axis=1)*2*(nx-1)*n*nz
        return real

    #def convertToRealPad(self, waveField):
    #    (nz,nx,n) = waveField.shape
    #    shiftField = np.fft.fftshift(waveField, axes=(0,2,))                        # Shift for easy padding
    #    shiftField[0,:,:] = 0.0                                                 # Zero out nyquist mode
    #    shiftField[:,:,0] = 0.0                                                 # Zero out nyquist mode
    #    padShiftField = np.pad(shiftField, ((int(nz/4), int(nz/4)),(0,int((nx-1)/2)),(int(n/4), int(n/4)))) # Pad zeros at front and back
    #    padField = np.fft.ifftshift(padShiftField, axes=(0,2,))                 # Shift back
    #
    #    realPad = np.fft.irfft(np.fft.ifft(np.fft.ifft(padField, axis=0), axis=2), axis=1)*(nx-1)*n*nz*27/8
    #    return realPad 
                                                    
    def plotPsuedocolor(self, x, y, z):
        YY, XX = np.meshgrid(y, x)
        fig, ax = plt.subplots()
        img = ax.pcolormesh(XX, YY, z, shading='gouraud')
        ax.set_aspect('equal')
        fig.colorbar(img)
        plt.show()

    def viewSliceByVar(self, var='u', index=4, axes=[0,1,2], pad=False): 
        if (pad):
            field = self.realPaddedFields[var]
        else:
            field = self.realFields[var]
        for direct in axes:
            view = field.take(indices=index, axis=direct)
            x1 = np.linspace(0, np.shape(view)[0]-1, np.shape(view)[0])
            x2 = np.linspace(0, np.shape(view)[1]-1, np.shape(view)[1])
            self.plotPsuedocolor(x1, x2, view)
            
    def viewFieldSlice(self, field, index=4, axes=[0,1,2], pad=False):
        for direct in axes:
            view = field.take(indices=index, axis=direct)
            x1 = np.linspace(0, np.shape(view)[0]-1, np.shape(view)[0])
            x2 = np.linspace(0, np.shape(view)[1]-1, np.shape(view)[1])
            self.plotPsuedocolor(x1, x2, view)        
                                                    
    def calculateRealCovarFields(self, fields=['rho','u','v','w']):
        for i in range(0, len(fields)):
            field1 = fields[i]
            for j in range(i, len(fields)):
                field2 = fields[j]
                name = field1 + field2
                self.realPaddedFields[name] = self.calculateRealCovar(field1, field2)
                                                    
    def calculateRealCovar(self, var1, var2):
        return np.multiply(self.realPaddedFields[var1],self.realPaddedFields[var2])

    def convertPaddedFieldsToWaveSpace(self, fields=None):
        if (fields):
            for key in fields:
                name = key + "hat"
                self.wavePaddedFields[name] = self.convertFieldToWavespace(self.realPaddedFields[key])        
        else:
            for key in self.realPaddedFields:
                name = key + "hat"
                self.wavePaddedFields[name] = self.convertFieldToWavespace(self.realPaddedFields[key])

    def convertFieldToWavespace(self, field):
        norm = field.shape[0] * field.shape[1] * field.shape[2]
        #hat = np.fft.fft(np.fft.fft(np.fft.fft(field, axis=0), axis=2), axis=1)/norm 
        hat = np.fft.rfft(np.fft.fft(np.fft.fft(field, axis=0), axis=2), axis=1)/norm  

        return hat
                                                    
    def filterPaddedWaveFields(self, fields=None, hFS=None, vFS=None):
        if (not hFS):
            hFS = 2 ** np.arange(0, np.log2(self.numPts[0])-2, 1)
        if (not vFS):
            vFS = 2 ** np.arange(0, np.log2(self.numPts[1])-2, 1)

        if (fields):
            for key in fields:
                name = key.split('hat')[0]
                #realField, waveField = self.fieldFilterDict( \
                #                                    self.wavePaddedFields[key], hFS, vFS, pad=True)
                #self.realFilteredFields[name] = realField  
                #self.waveFilteredFields[name] = waveField 
                self.realFilteredFields[name] = self.fieldFilterDict( \
                                                    self.wavePaddedFields[key], hFS, vFS, pad=True)
        else:
            for key in self.wavePaddedFields:
                name = key.split('hat')[0]
                #realField, waveField = self.fieldFilterDict( \
                #                                    self.wavePaddedFields[key], hFS, vFS, pad=True)  
                #self.realFilteredFields[name] = realField  
                #self.waveFilteredFields[name] = waveField   
                self.realFilteredFields[name] = self.fieldFilterDict( \
                                                    self.wavePaddedFields[key], hFS, vFS, pad=True) 

    def filterWaveFields(self, fields=None, hFS=None, vFS=None):
        if (not hFS):
            hFS = 2 ** np.arange(0, np.log2(self.numPts[0])-2, 1)
        if (not vFS):
            vFS = 2 ** np.arange(0, np.log2(self.numPts[1])-2, 1)

        if (fields):
            for key in fields:
                name = key.split('hat')[0]
                self.realFilteredFields[name] = self.fieldFilterDict( \
                                                    self.waveFields[key], hFS, vFS, pad=False)        
        else:
            for key in self.waveFields:
                name = key.split('hat')[0]
                self.realFilteredFields[name] = self.fieldFilterDict( \
                                                    self.waveFields[key], hFS, vFS, pad=False) 
                
    def applySharpSpectralFilter(self, field, filtH=1, filtV=1, padded=True):
        # field: wavespace representation of variable (kz, kx, ky)
        # filtH: ratio of horizontal (kx,kz) filter to grid (1 implies no filter applied)
        # filtV: ratio of vertical (ky) filter to grid (1 implies no filter applied)
        # pad: whether the field is 3/2 padded or not
        
        # shift field for easy filtering
        shiftField = np.fft.fftshift(field, axes=(0,2,)) 

        # number of grid points in horizontal and vertical directions 
        # (multiply by 2/3 if field is padded)
        if (padded):
            nH = int(field.shape[0]*2/3)
            nV = int(field.shape[2]*2/3)
        else:
            nH = int(field.shape[0])
            nV = int(field.shape[2])

        # wavenumbers are now [-3/2*nyq, -3/2*(nyq-1), ..., -2, -1, 0, 1, 2, ..., 3/2(nyq-1)] in each direction
        # so to filter we will only take the middle wavenumbers <= cutoff
        # this will remove the padding and apply the filter
        horizLowK  = int(field.shape[0]/2 - nH/(2*filtH))
        horizHighK = int(field.shape[0]/2 + nH/(2*filtH))
        vertLowK   = int(field.shape[2]/2 - nV/(2*filtV))
        vertHighK  = int(field.shape[2]/2 + nV/(2*filtV))  

        filtField = shiftField[horizLowK:horizHighK, :int(nH/filtH/2+1), vertLowK:vertHighK]

        # now since we filtered, the filtered array's nyquist values need to be removed since they
        # may contain actual information that we can't resolve
        filtField[0,:,:] = 0.0
        filtField[:,-1,:] = 0.0
        filtField[:,:,0] = 0.0

        # now shift back so that the field is laid out properly for fft
        filtField = np.fft.ifftshift(filtField, axes=(0,2,))
        
        return filtField
        
#    def realFilter(self, field, filtH=1, filtV=1, padded=True):
#        # field: wavespace representation of variable (kz, kx, ky)
#        # filtH: ratio of horizontal (kx,kz) filter to grid (1 implies no filter applied)
#        # filtV: ratio of vertical (ky) filter to grid (1 implies no filter applied)
#        # pad: whether the field is 3/2 padded or not
#
#        # shift field for easy filtering
#        shiftField = np.fft.fftshift(field, axes=(0,1,2)) 
#
#        # number of grid points in horizontal and vertical directions 
#        # (multiply by 2/3 if field is padded)
#        if (padded):
#            nH = int(field.shape[0]*2/3)
#            nV = int(field.shape[2]*2/3)
#        else:
#            nH = int(field.shape[0])
#            nV = int(field.shape[2])
#
#        # wavenumbers are now [-3/2*nyq, -3/2*(nyq-1), ..., -2, -1, 0, 1, 2, ..., 3/2(nyq-1)] in each direction
#        # so to filter we will only take the middle wavenumbers <= cutoff
#        # this will remove the padding and apply the filter
#        horizLowK  = int(field.shape[0]/2 - nH/(2*filtH))
#        horizHighK = int(field.shape[0]/2 + nH/(2*filtH))
#        vertLowK   = int(field.shape[2]/2 - nV/(2*filtV))
#        vertHighK  = int(field.shape[2]/2 + nV/(2*filtV))  
#
#        filtField = shiftField[horizLowK:horizHighK, horizLowK:horizHighK, vertLowK:vertHighK]
#
#        # now since we filtered, the filtered array's nyquist values need to be removed since they
#        # may contain actual information that we can't resolve
#        filtField[0,:,:] = 0.0
#        filtField[:,0,:] = 0.0
#        filtField[:,:,0] = 0.0
#
#        # now shift back so that the field is laid out properly for fft
#        filtField = np.fft.ifftshift(filtField, axes=(0,1,2))
#        
#        # inverse fft
#        norm = filtField.shape[0]*filtField.shape[1]*filtField.shape[2]
#        realField = np.fft.ifft(np.fft.ifft(np.fft.ifft(filtField, axis=0), axis=2), axis=1)*norm
#        if (np.max(np.imag(realField)) > 1.0e15):
#            print (np.max(np.imag(realField)), " too large for imaginary component of a real fft?")
#
#        return np.real(realField) #, filtField

    def fieldFilterDict(self, field, horizFilterScales, vertFilterScales, pad=True):
        returnDict = {}
        #waveDict = {}
        for horizFilter in horizFilterScales:
            for vertFilter in vertFilterScales:
                name = str(int(horizFilter)) + "," + str(int(vertFilter))
                #realField, waveField = self.realFilter(field, filtH=horizFilter, filtV=vertFilter, padded=pad)
                filtField = self.applySharpSpectralFilter(field, filtH=horizFilter, filtV=vertFilter, padded=pad)
                returnDict[name] = self.convertToReal(filtField)
                #returnDict[name] = self.realFilter(field, filtH=horizFilter, filtV=vertFilter, padded=pad)
                #returnDict[name] = realField
                #waveDict[name] = waveField

        return returnDict #, waveDict
    
    def addGradToFields(self, fields=['uhat','vhat','what','rhohat'], hFS=None, vFS=None):
        for key in fields:
            #ddxhat, ddyhat, ddzhat = self.gradField(self.wavePaddedFields[key])
            ddxhat, ddyhat, ddzhat = self.gradField(self.waveFields[key])
            nameX = "d" + key.split("hat")[0] + "dxhat"
            self.waveFields[nameX] = ddxhat
            nameY = "d" + key.split("hat")[0] + "dyhat"
            self.waveFields[nameY] = ddyhat
            nameZ = "d" + key.split("hat")[0] + "dzhat"
            self.waveFields[nameZ] = ddzhat

            self.filterWaveFields(fields=[nameX, nameY, nameZ], hFS=hFS, vFS=vFS)

#    def gradField(self, padField):
    def gradField(self, waveField):
#        shiftField = np.fft.fftshift(padField, axes=(0,1,2)) 
#        horizLowK  = int(padField.shape[0]*(1/6))
#        horizHighK = int(padField.shape[0]*(5/6))
#        vertLowK   = int(padField.shape[2]*(1/6))
#        vertHighK  = int(padField.shape[2]*(5/6))  
#
#        filtField = shiftField[horizLowK:horizHighK, horizLowK:horizHighK, vertLowK:vertHighK]
#        filtField[0,:,:] = 0.0
#        filtField[:,0,:] = 0.0
#        filtField[:,:,0] = 0.0
#        field = np.fft.ifftshift(filtField, axes=(0,1,2))    
#
#        kh = field.shape[0]
#        kv = field.shape[2]
#
        kx, ky, kz = self.calcWeights()

        gradXhat = np.multiply(waveField,kx*(0+1.j))
        gradYhat = np.multiply(waveField,ky*(0+1.j))
        gradZhat = np.multiply(waveField,kz*(0+1.j))

        return gradXhat, gradYhat, gradZhat
    
    def calcWeights(self):
        nx = self.numPts[0]
        ny = self.numPts[1]
        nz = self.numPts[2]
        yBoxScale = self.yScale
        
        #kxFxn = lambda k, j, i: j % (nx/2) - np.floor(2*j/nx)*nx/2
        #kx = np.fromfunction(np.vectorize(kxFxn), (int(nz), int(nx), int(ny)), dtype=float)
        kxFxn = lambda k, j, i: j
        kx = np.fromfunction(np.vectorize(kxFxn), (int(nz), int(nx/2+1), int(ny)), dtype=float)

        kyFxn = lambda k, j, i: (i/yBoxScale) % (nx/2) - np.floor(2*i/(nx*yBoxScale))*nx/2
        #ky = np.fromfunction(np.vectorize(kyFxn), (int(nz), int(nx), int(ny)), dtype=float)
        ky = np.fromfunction(np.vectorize(kyFxn), (int(nz), int(nx/2+1), int(ny)), dtype=float)

        kzFxn = lambda k, j, i: k % (nz/2) - np.floor(2*k/nz)*nz/2
        #kz = np.fromfunction(np.vectorize(kzFxn), (int(nz), int(nx), int(ny)), dtype=float)
        kz = np.fromfunction(np.vectorize(kzFxn), (int(nz), int(nx/2+1), int(ny)), dtype=float)

        return kx, ky, kz
    
    def savePickle(self):
        f = open(self.pickleName, 'wb')
        pickle.dump(self.__dict__, f, 2)
        f.close()
        
    def calcTauExact(self):
        self.realFilteredFields['tauExactUU'] = {}
        self.realFilteredFields['tauExactUV'] = {}
        self.realFilteredFields['tauExactUW'] = {}
        self.realFilteredFields['tauExactVV'] = {}
        self.realFilteredFields['tauExactVW'] = {}
        self.realFilteredFields['tauExactWW'] = {}
        self.realFilteredFields['tauModelUU'] = {}
        self.realFilteredFields['tauModelUV'] = {}
        self.realFilteredFields['tauModelUW'] = {}
        self.realFilteredFields['tauModelVV'] = {}
        self.realFilteredFields['tauModelVW'] = {}
        self.realFilteredFields['tauModelWW'] = {}
        self.realFilteredFields['qExactRU'] = {}
        self.realFilteredFields['qExactRV'] = {}
        self.realFilteredFields['qExactRW'] = {}
        self.realFilteredFields['qExactRR'] = {}
        for key in self.realFilteredFields['u'].keys():   # different filter levels
            self.realFilteredFields['tauExactUU'][key] = self.calcRealTauDiff('uu', 'u', 'u', key)
            self.realFilteredFields['tauExactUV'][key] = self.calcRealTauDiff('uv', 'u', 'v', key)
            self.realFilteredFields['tauExactUW'][key] = self.calcRealTauDiff('uw', 'u', 'w', key)
            self.realFilteredFields['tauExactVV'][key] = self.calcRealTauDiff('vv', 'v', 'v', key)
            self.realFilteredFields['tauExactVW'][key] = self.calcRealTauDiff('vw', 'v', 'w', key)
            self.realFilteredFields['tauExactWW'][key] = self.calcRealTauDiff('ww', 'w', 'w', key)
            self.realFilteredFields['tauModelUU'][key] = self.calcRealSijComp('dudx', 'dudx', key)
            self.realFilteredFields['tauModelUV'][key] = self.calcRealSijComp('dudy', 'dvdx', key)
            self.realFilteredFields['tauModelUW'][key] = self.calcRealSijComp('dudz', 'dwdx', key)
            self.realFilteredFields['tauModelVV'][key] = self.calcRealSijComp('dvdy', 'dvdy', key)
            self.realFilteredFields['tauModelVW'][key] = self.calcRealSijComp('dvdz', 'dwdy', key)
            self.realFilteredFields['tauModelWW'][key] = self.calcRealSijComp('dwdz', 'dwdz', key)
            self.realFilteredFields['qExactRU'][key] = self.calcRealTauDiff('rhou', 'rho', 'u', key)
            self.realFilteredFields['qExactRV'][key] = self.calcRealTauDiff('rhov', 'rho', 'v', key)
            self.realFilteredFields['qExactRW'][key] = self.calcRealTauDiff('rhow', 'rho', 'w', key)
            self.realFilteredFields['qExactRR'][key] = self.calcRealTauDiff('rhorho', 'rho', 'rho', key)

    def calcRealTauDiff(self, filtProdKey, filtCompKey1, filtCompKey2, filtLevel):
        filtProd = self.realFilteredFields[filtProdKey][filtLevel]
        prodFilt = np.multiply(self.realFilteredFields[filtCompKey1][filtLevel],self.realFilteredFields[filtCompKey2][filtLevel])
        return filtProd - prodFilt    

    def calcRealSijComp(self, filtCompKey1, filtCompKey2, filtLevel):
        return self.realFilteredFields[filtCompKey1][filtLevel] + self.realFilteredFields[filtCompKey1][filtLevel]
```

```python
field = xhat
filtH = 1
filtV = 1
# field: 3/2 padded wavespace representation of variable (kz, kx, ky)
# filtH: ratio of horizontal (kx,kz) filter to grid (1 implies no filter applied)
# filtV: ratio of vertical (ky) filter to grid (1 implies no filter applied)

# shift field for easy filtering
shiftField = np.fft.fftshift(field, axes=(0,1,2)) 

# number of grid points in horizontal and vertical directions 
# (multiply by 2/3 since field is padded)
nH = int(field.shape[0]*2/3)
nV = int(field.shape[2]*2/3)

# wavenumbers are now [-3/2*nyq, -3/2*(nyq-1), ..., -2, -1, 0, 1, 2, ..., 3/2(nyq-1)] in each direction
# so to filter we will only take the middle wavenumbers <= cutoff
# this will remove the padding and apply the filter
horizLowK  = int(field.shape[0]*(3*filtH - 2) / (6 * filtH))
horizHighK = int(field.shape[0]*(3*filtH + 2) / (6 * filtH))
vertLowK   = int(field.shape[2]*(3*filtV - 2) / (6 * filtV))
vertHighK  = int(field.shape[2]*(3*filtV + 2) / (6 * filtV))    
filtField = shiftField[horizLowK:horizHighK, horizLowK:horizHighK, vertLowK:vertHighK]

# now since we filtered, the filtered array's nyquist values need to be removed since they
# may contain actual information that we can't resolve
filtField[0,:,:] = 0.0
filtField[:,0,:] = 0.0
filtField[:,:,0] = 0.0

# now shift back so that the field is laid out properly for fft
filtField = np.fft.ifftshift(filtField, axes=(0,1,2))

# inverse fft
norm = filtField.shape[0]*filtField.shape[1]*filtField.shape[2]
realField = np.fft.ifft(np.fft.ifft(np.fft.ifft(filtField, axis=0), axis=2), axis=1)*norm
if (np.max(np.imag(realField)) > 1.0e15):
    print (np.max(np.imag(realField)), " too large for imaginary component of a real fft?")
```

```python
import numpy as np
import h5py as hdf
import matplotlib as mpl
import matplotlib.pyplot as plt
import pickle

class oldStratFile:
    def __init__(self, filepath):
        self.path = filepath
        self.pickled = False
        self.checkForPickleDump()
        self.waveFields = {}         # z, x, y
        self.wavePaddedFields = {}   # z, x, y
        self.waveFilteredFields = {}
        self.realFields = {}
        self.realPaddedFields = {}   # z, x, y
        self.realFilteredFields = {}
        self.processHDF5()

    def checkForPickleDump(self):
        self.pickled = False
        
    def processHDF5(self):
        hdfData = hdf.File(self.path,"r")
        self.numPts = [hdfData.get('nx')[()], hdfData.get('ny')[()], hdfData.get('nz')[()]]
        self.time = hdfData.get('time')[()]
        self.yScale = hdfData.get('LMFy')[()]
        self.visc = hdfData.get('nu')[()]
        self.diff = hdfData.get('diff')[()]
        self.grav = hdfData.get('grav')[()]
        self.gradv = hdfData.get('gradv')[()]
        self.step = hdfData.get('timestep')[()]
        
        bruVais = str(int(np.sqrt(self.grav)))
        self.pickleName = "N" + bruVais + "-" + str(int(self.numPts[0])) + str(int(self.numPts[1])) \
                              + "-" + str(int(self.step)) + ".p"

        self.processHDFFields(hdfData)
        
    def processHDFFields(self, hdfData):
        # Something that checks for a pickle dump
        # self.pickled = search directory for pickle file for 
        # if (self.pickled):
            # read it in
        # else:
            self.readHDFFields(hdfData)
            #self.generateInitialRealFields()
            self.generateInitialRealPaddedFields()
            
    def readHDFFields(self, hdfData, fields=['rho','u','v','w']):
        for field in fields:
            fieldData = hdfData[field][()]
            nz,nx,n,_ = fieldData.shape
            wavespace = np.zeros((nz,nx+1,n), dtype=complex)
            wavespace[:,:nx,:].real = fieldData[:,:,:,0]
            wavespace[:,:nx,:].imag = fieldData[:,:,:,1]
            name = field + 'hat'
            self.waveFields[name] = wavespace   

    def generateRealFields(self, fields=['rhohat','uhat','vhat','what']):
        for key in fields:
            name = key[:-3]    # drop the _hat from the waveField name
            self.realFields[name] = self.convertToReal(self.waveFields[key])
            
    def generateInitialRealPaddedFields(self):
        for key, waveField in self.waveFields.items():
            name = key[:-3]    # drop the _hat from the waveField name
            self.realPaddedFields[name] = self.convertToRealPad(waveField)

    def convertToReal(self, waveField):
        (nz,nx,n) = waveField.shape
        real = np.fft.irfft(np.fft.ifft(np.fft.ifft(waveField, axis=0), axis=2), axis=1)*(nx-1)*n*nz
        return real

    def convertToRealPad(self, waveField):
        (nz,nx,n) = waveField.shape
        shiftField = np.fft.fftshift(waveField, axes=(0,2,))                        # Shift for easy padding
        shiftField[0,:,:] = 0.0                                                 # Zero out nyquist mode
        shiftField[:,:,0] = 0.0                                                 # Zero out nyquist mode
        padShiftField = np.pad(shiftField, ((int(nz/4), int(nz/4)),(0,int((nx-1)/2)),(int(n/4), int(n/4)))) # Pad zeros at front and back
        padField = np.fft.ifftshift(padShiftField, axes=(0,2,))                 # Shift back

        realPad = np.fft.irfft(np.fft.ifft(np.fft.ifft(padField, axis=0), axis=2), axis=1)*(nx-1)*n*nz*27/8
        return realPad 
                                                    
    def plotPsuedocolor(self, x, y, z):
        YY, XX = np.meshgrid(y, x)
        fig, ax = plt.subplots()
        img = ax.pcolormesh(XX, YY, z, shading='gouraud')
        ax.set_aspect('equal')
        fig.colorbar(img)
        plt.show()

    def viewSliceByVar(self, var='u', index=4, axes=[0,1,2], pad=False): 
        if (pad):
            field = self.realPaddedFields[var]
        else:
            field = self.realFields[var]
        for direct in axes:
            view = field.take(indices=index, axis=direct)
            x1 = np.linspace(0, np.shape(view)[0]-1, np.shape(view)[0])
            x2 = np.linspace(0, np.shape(view)[1]-1, np.shape(view)[1])
            self.plotPsuedocolor(x1, x2, view)
            
    def viewFieldSlice(self, field, index=4, axes=[0,1,2], pad=False):
        for direct in axes:
            view = field.take(indices=index, axis=direct)
            x1 = np.linspace(0, np.shape(view)[0]-1, np.shape(view)[0])
            x2 = np.linspace(0, np.shape(view)[1]-1, np.shape(view)[1])
            self.plotPsuedocolor(x1, x2, view)        
                                                    
    def calculateRealCovarFields(self, fields=['rho','u','v','w']):
        for i in range(0, len(fields)):
            field1 = fields[i]
            for j in range(i, len(fields)):
                field2 = fields[j]
                name = field1 + field2
                self.realPaddedFields[name] = self.calculateRealCovar(field1, field2)
                                                    
    def calculateRealCovar(self, var1, var2):
        return np.multiply(self.realPaddedFields[var1],self.realPaddedFields[var2])

    def convertPaddedFieldsToWaveSpace(self, fields=None):
        if (fields):
            for key in fields:
                name = key + "hat"
                self.wavePaddedFields[name] = self.convertFieldToWavespace(self.realPaddedFields[key])        
        else:
            for key in self.realPaddedFields:
                name = key + "hat"
                self.wavePaddedFields[name] = self.convertFieldToWavespace(self.realPaddedFields[key])

    def convertFieldToWavespace(self, field):
        norm = field.shape[0] * field.shape[1] * field.shape[2]
        hat = np.fft.fft(np.fft.fft(np.fft.fft(field, axis=0), axis=2), axis=1)/norm  

        return hat
                                                    
    def filterPaddedWaveFields(self, fields=None, hFS=None, vFS=None):
        if (not hFS):
            hFS = 2 ** np.arange(0, np.log2(self.numPts[0])-2, 1)
        if (not vFS):
            vFS = 2 ** np.arange(0, np.log2(self.numPts[1])-2, 1)

        if (fields):
            for key in fields:
                name = key.split('hat')[0]
                #realField, waveField = self.fieldFilterDict( \
                #                                    self.wavePaddedFields[key], hFS, vFS, pad=True)
                #self.realFilteredFields[name] = realField  
                #self.waveFilteredFields[name] = waveField 
                self.realFilteredFields[name] = self.fieldFilterDict( \
                                                    self.wavePaddedFields[key], hFS, vFS, pad=True)
        else:
            for key in self.wavePaddedFields:
                name = key.split('hat')[0]
                #realField, waveField = self.fieldFilterDict( \
                #                                    self.wavePaddedFields[key], hFS, vFS, pad=True)  
                #self.realFilteredFields[name] = realField  
                #self.waveFilteredFields[name] = waveField   
                self.realFilteredFields[name] = self.fieldFilterDict( \
                                                    self.wavePaddedFields[key], hFS, vFS, pad=True) 

    def filterWaveFields(self, fields=None, hFS=None, vFS=None):
        if (not hFS):
            hFS = 2 ** np.arange(0, np.log2(self.numPts[0])-2, 1)
        if (not vFS):
            vFS = 2 ** np.arange(0, np.log2(self.numPts[1])-2, 1)

        if (fields):
            for key in fields:
                name = key.split('hat')[0]
                self.realFilteredFields[name] = self.fieldFilterDict( \
                                                    self.waveFields[key], hFS, vFS, pad=False)        
        else:
            for key in self.waveFields:
                name = key.split('hat')[0]
                self.realFilteredFields[name] = self.fieldFilterDict( \
                                                    self.waveFields[key], hFS, vFS, pad=False) 

    def realFilter(self, field, filtH=1, filtV=1, padded=True):
        # field: wavespace representation of variable (kz, kx, ky)
        # filtH: ratio of horizontal (kx,kz) filter to grid (1 implies no filter applied)
        # filtV: ratio of vertical (ky) filter to grid (1 implies no filter applied)
        # pad: whether the field is 3/2 padded or not

        # shift field for easy filtering
        shiftField = np.fft.fftshift(field, axes=(0,1,2)) 

        # number of grid points in horizontal and vertical directions 
        # (multiply by 2/3 if field is padded)
        if (padded):
            nH = int(field.shape[0]*2/3)
            nV = int(field.shape[2]*2/3)
        else:
            nH = int(field.shape[0])
            nV = int(field.shape[2])

        # wavenumbers are now [-3/2*nyq, -3/2*(nyq-1), ..., -2, -1, 0, 1, 2, ..., 3/2(nyq-1)] in each direction
        # so to filter we will only take the middle wavenumbers <= cutoff
        # this will remove the padding and apply the filter
        horizLowK  = int(field.shape[0]/2 - nH/(2*filtH))
        horizHighK = int(field.shape[0]/2 + nH/(2*filtH))
        vertLowK   = int(field.shape[2]/2 - nV/(2*filtV))
        vertHighK  = int(field.shape[2]/2 + nV/(2*filtV))  

        filtField = shiftField[horizLowK:horizHighK, horizLowK:horizHighK, vertLowK:vertHighK]

        # now since we filtered, the filtered array's nyquist values need to be removed since they
        # may contain actual information that we can't resolve
        filtField[0,:,:] = 0.0
        filtField[:,0,:] = 0.0
        filtField[:,:,0] = 0.0

        # now shift back so that the field is laid out properly for fft
        filtField = np.fft.ifftshift(filtField, axes=(0,1,2))
        
        # inverse fft
        norm = filtField.shape[0]*filtField.shape[1]*filtField.shape[2]
        realField = np.fft.ifft(np.fft.ifft(np.fft.ifft(filtField, axis=0), axis=2), axis=1)*norm
        if (np.max(np.imag(realField)) > 1.0e15):
            print (np.max(np.imag(realField)), " too large for imaginary component of a real fft?")

        return np.real(realField) #, filtField


    def fieldFilterDict(self, field, horizFilterScales, vertFilterScales, pad=True):
        returnDict = {}
        #waveDict = {}
        for horizFilter in horizFilterScales:
            for vertFilter in vertFilterScales:
                name = str(int(horizFilter)) + "," + str(int(vertFilter))
                #realField, waveField = self.realFilter(field, filtH=horizFilter, filtV=vertFilter, padded=pad)
                returnDict[name] = self.realFilter(field, filtH=horizFilter, filtV=vertFilter, padded=pad)
                #returnDict[name] = realField
                #waveDict[name] = waveField

        return returnDict #, waveDict
    
    def addGradToFields(self, fields=['uhat','vhat','what','rhohat'], hFS=None, vFS=None):
        for key in fields:
            ddxhat, ddyhat, ddzhat = self.gradField(self.wavePaddedFields[key])
            nameX = "d" + key.split("hat")[0] + "dxhat"
            self.waveFields[nameX] = ddxhat
            nameY = "d" + key.split("hat")[0] + "dyhat"
            self.waveFields[nameY] = ddyhat
            nameZ = "d" + key.split("hat")[0] + "dzhat"
            self.waveFields[nameZ] = ddzhat

            self.filterWaveFields(fields=[nameX, nameY, nameZ], hFS=hFS, vFS=vFS)

    def gradField(self, padField):
        shiftField = np.fft.fftshift(padField, axes=(0,1,2)) 
        horizLowK  = int(padField.shape[0]*(1/6))
        horizHighK = int(padField.shape[0]*(5/6))
        vertLowK   = int(padField.shape[2]*(1/6))
        vertHighK  = int(padField.shape[2]*(5/6))  

        filtField = shiftField[horizLowK:horizHighK, horizLowK:horizHighK, vertLowK:vertHighK]
        filtField[0,:,:] = 0.0
        filtField[:,0,:] = 0.0
        filtField[:,:,0] = 0.0
        field = np.fft.ifftshift(filtField, axes=(0,1,2))    

        kh = field.shape[0]
        kv = field.shape[2]

        kx, ky, kz = self.calcWeights()

        gradXhat = np.multiply(field,kx*(0+1.j))
        gradYhat = np.multiply(field,ky*(0+1.j))
        gradZhat = np.multiply(field,kz*(0+1.j))

        return gradXhat, gradYhat, gradZhat
    
    def calcWeights(self):
        nx = self.numPts[0]
        ny = self.numPts[1]
        nz = self.numPts[2]
        yBoxScale = self.yScale
        
        kxFxn = lambda k, j, i: j % (nx/2) - np.floor(2*j/nx)*nx/2
        kx = np.fromfunction(np.vectorize(kxFxn), (int(nz), int(nx), int(ny)), dtype=float)

        kyFxn = lambda k, j, i: (i/yBoxScale) % (nx/2) - np.floor(2*i/(nx*yBoxScale))*nx/2
        ky = np.fromfunction(np.vectorize(kyFxn), (int(nz), int(nx), int(ny)), dtype=float)

        kzFxn = lambda k, j, i: k % (nz/2) - np.floor(2*k/nz)*nz/2
        kz = np.fromfunction(np.vectorize(kzFxn), (int(nz), int(nx), int(ny)), dtype=float)

        return kx, ky, kz
    
    def savePickle(self):
        f = open(self.pickleName, 'wb')
        pickle.dump(self.__dict__, f, 2)
        f.close()
        
    def calcTauExact(self):
        self.realFilteredFields['tauExactUU'] = {}
        self.realFilteredFields['tauExactUV'] = {}
        self.realFilteredFields['tauExactUW'] = {}
        self.realFilteredFields['tauExactVV'] = {}
        self.realFilteredFields['tauExactVW'] = {}
        self.realFilteredFields['tauExactWW'] = {}
        self.realFilteredFields['tauModelUU'] = {}
        self.realFilteredFields['tauModelUV'] = {}
        self.realFilteredFields['tauModelUW'] = {}
        self.realFilteredFields['tauModelVV'] = {}
        self.realFilteredFields['tauModelVW'] = {}
        self.realFilteredFields['tauModelWW'] = {}
        self.realFilteredFields['qExactRU'] = {}
        self.realFilteredFields['qExactRV'] = {}
        self.realFilteredFields['qExactRW'] = {}
        self.realFilteredFields['qExactRR'] = {}
        for key in self.realFilteredFields['u'].keys():   # different filter levels
            self.realFilteredFields['tauExactUU'][key] = self.calcRealTauDiff('uu', 'u', 'u', key)
            self.realFilteredFields['tauExactUV'][key] = self.calcRealTauDiff('uv', 'u', 'v', key)
            self.realFilteredFields['tauExactUW'][key] = self.calcRealTauDiff('uw', 'u', 'w', key)
            self.realFilteredFields['tauExactVV'][key] = self.calcRealTauDiff('vv', 'v', 'v', key)
            self.realFilteredFields['tauExactVW'][key] = self.calcRealTauDiff('vw', 'v', 'w', key)
            self.realFilteredFields['tauExactWW'][key] = self.calcRealTauDiff('ww', 'w', 'w', key)
            self.realFilteredFields['tauModelUU'][key] = self.calcRealSijComp('dudx', 'dudx', key)
            self.realFilteredFields['tauModelUV'][key] = self.calcRealSijComp('dudy', 'dvdx', key)
            self.realFilteredFields['tauModelUW'][key] = self.calcRealSijComp('dudz', 'dwdx', key)
            self.realFilteredFields['tauModelVV'][key] = self.calcRealSijComp('dvdy', 'dvdy', key)
            self.realFilteredFields['tauModelVW'][key] = self.calcRealSijComp('dvdz', 'dwdy', key)
            self.realFilteredFields['tauModelWW'][key] = self.calcRealSijComp('dwdz', 'dwdz', key)
            self.realFilteredFields['qExactRU'][key] = self.calcRealTauDiff('rhou', 'rho', 'u', key)
            self.realFilteredFields['qExactRV'][key] = self.calcRealTauDiff('rhov', 'rho', 'v', key)
            self.realFilteredFields['qExactRW'][key] = self.calcRealTauDiff('rhow', 'rho', 'w', key)
            self.realFilteredFields['qExactRR'][key] = self.calcRealTauDiff('rhorho', 'rho', 'rho', key)

    def calcRealTauDiff(self, filtProdKey, filtCompKey1, filtCompKey2, filtLevel):
        filtProd = self.realFilteredFields[filtProdKey][filtLevel]
        prodFilt = np.multiply(self.realFilteredFields[filtCompKey1][filtLevel],self.realFilteredFields[filtCompKey2][filtLevel])
        return filtProd - prodFilt    

    def calcRealSijComp(self, filtCompKey1, filtCompKey2, filtLevel):
        return self.realFilteredFields[filtCompKey1][filtLevel] + self.realFilteredFields[filtCompKey1][filtLevel]
```
